## Ejercicio 1: Condición de resonancia

Al someter un material magnético a un campo magnético y a ondas microondas, los espines del material precesan de acuerdo con la ecuación:

$$sin\left(2\theta_M \right)=\frac{2H_R}{4\pi M_{eff}}sin(\theta_M-\theta_H)$$

con $4\pi M_{eff}=1.7$ y $H_R$ y $\theta_H$.
Solucione la ecuación para $\theta_M$ por el método de bisección y otro de su elección. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Metodo de Bisección


```python
def Biseccion(f,a,b):
    
    #se define un valor medio y se evalua la función en los valores deseados
    r=(a+b)/2
    f_a=f(a)
    f_b=f(b)
    f_r=f(r)
    an,bn,rn=a,b,r
    
    while np.abs(f_r)>1e-12:
        
        #se termina el proceso si alguno de los valores a,b o r son raíz
        if f_r*f_a==0 or f_r*f_b==0: 
            break
            
        #según la condición dada de se reemplaza el alguno de los límites por el valor medio    
        if f_r*f_a<0: 
            an=an
            bn=rn
        elif f_r*f_a>0:
            an=rn
            bn=bn 
        
        #según el resultados se vuelve a evaluar la función hasta que se acerque 
        #lo suficiente a cero
        rn=(an+bn)/2
        f_r=f(rn)
        f_a=f(an)
        f_b=f(bn)
        
    return rn     
    
```

```python
angle=np.arange(0,95,5)
rad_angle=np.pi*(1/180)*angle

H_r=np.array([2.3179,2.3026,2.3319,2.3653,2.4345,2.5137,2.6138,
              2.7601,2.9222,3.0981,3.3646,3.6393,3.9168,4.332,
              4.6016,4.9356,5.2205,5.3954,5.4483])
```

```python
#se define la función a solucionar
def cond_res(x):
    Meff=1.7
    sol=2*H_R*(1/Meff)*np.sin(x-ang)-np.sin(2*x)
    return sol


sol=[]

#se encuentra la raíz para cada para angulo y H_r
for i in range(len(angle)):
    ang=rad_angle[i]
    H_R=H_r[i]
    x=Biseccion(cond_res,0,np.pi/2)
    sol.append(x)
    
deg_sol=(180/np.pi)*np.array(sol) 
      
```

```python

plt.figure(dpi=100)
plt.plot(angle,deg_sol)
plt.scatter(angle,deg_sol)
plt.xlabel('$ \\theta_H $°')
plt.ylabel('$ \\theta_M $°')
plt.title('Solución con Bisección')
plt.grid()
plt.show()
```

### Método de Newton-Raphson

```python
#se define la función y su derivada que fue encontrada de manera analítica
def cond_res(x):
    Meff=1.7
    sol=2*H_R*(1/Meff)*np.sin(x-ang)-np.sin(2*x)
    return sol
def dcond_res(x):
    Meff=1.7
    sol=2*H_R*(1/Meff)*np.cos(x-ang)-2*np.cos(2*x)
    return sol

def NR(f,df,x0):
    fx=f(x0)
    dfx=df(x0)
    xn=x0
    #en un condicional de define la tangente de la función de tal manera que iterando 
    #se llegue a la raíz
    while np.abs(fx)>1e-12:
        if f(x0)==0: break
        xn=xn-fx/dfx
        fx=f(xn)
        dfx=df(xn)
        #print('z')
    return xn    
```

```python
#se calcula la función para los ángulos y los H_R respectivos 
soln=[]
for i in range(len(angle)):
    ang=rad_angle[i]
    H_R=H_r[i]
    x=NR(cond_res,dcond_res,np.pi/2)
    soln.append(x)
deg_soln=180/(np.pi)*np.array(soln)

```

```python
plt.figure(dpi=100)
plt.plot(angle,deg_soln)
plt.scatter(angle,deg_soln)
plt.xlabel('$ \\theta_H $°')
plt.ylabel('$ \\theta_M $°')
plt.title('Solución con Newton-Raphson')
plt.grid()
plt.show()
```

```python
plt.figure(dpi=100)
plt.plot(angle,deg_sol,label='Método de bisección')
plt.plot(angle,deg_soln,label='Método de Newton-Raphson')
plt.xlabel('$ \\theta_H $°')
plt.ylabel('$ \\theta_M $°')
plt.title('Comparación de los métodos de Newton-Raphson y de Bisección')
plt.grid()
plt.legend()
plt.show()
```
